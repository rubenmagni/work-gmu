```python
def save_list(input_list, file_name):
    """A list (input_list) is saved in a file (file_name)"""
    with open(file_name, 'w') as fh:
        for item in input_list:
            fh.write('{0}\n'.format(item))
    return None
```

```python
import xlwt
list1 = [1,2,3,4,5]
list2 = [234,267,281,301,331]
wb = xlwt.Workbook()
ws = wb.add_sheet('First sheet')
ws.write(0,0,'Column A')
ws.write(0,1,'Column B')
i = 1
for x,y in zip(list1,list2): #Walk two list at the same time.
    ws.write(i,0,x) # Row, Column, Data.
    ws.write(i,1,y)
    i += 1
wb.save('mynewfile.xls')
```

```python
# heatmap for sepsis
heatmap = pd.read_csv("for_heatmap.csv")
```

```python
ct_counts = heatmap.groupby(['Sample name', 'Genus']).size() # I use groupby and size to find the countsize of the combination of the two variable levels into a data series
ct_counts = ct_counts.reset_index(name = 'count') # then I use reset index to convert the series into a dataframe
ct_counts = ct_counts.pivot(index = 'Sample name', columns = 'Genus', values = 'count')
```
